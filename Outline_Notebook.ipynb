# Campaign Expenditures and Congressional Race Results in North Carolina



---



### Purpose of the Research
This research aims to examine how campaign expenditure influences the outcomes of congressional contests in North Carolina. We want to see if there is a clear link between the amount of money spent and the success rate of candidates. We seek to discover patterns that can influence future campaign strategies by analyzing precise spending data from the Federal Election Commission (FEC) and comparing them to election outcomes from the North Carolina State Board of Elections and essentially, asking whether spending more money leads to better outcomes in North Carolina's congressional races.

The findings can help political groups make more informed resource allocation decisions. Campaigns can engage voters more effectively if they uncover effective spending techniques. Furthermore, the findings can assist politicians in developing legislation that guarantees fair elections. Political parties can utilize this information to target competitive districts better. Overall, I intend to offer actionable findings that will improve the fairness and effectiveness of North Carolina's electoral process.


### Hypothesis Tested

This study tests three main hypotheses about how campaign spending impacts congressional races in North Carolina:

1. **Hypothesis 1** (H1): Campaign spending correlates positively with vote percentages. In other words, candidates spending more money on their campaigns tend to receive more votes. The assumption is that more expenditure boosts a candidate's exposure and outreach, resulting in more voters.


2. **Hypothesis 2** (H2): Higher campaign spending dramatically raises the likelihood of winning a congressional seat. This shows that increased expenditure correlates with a larger vote share and directly contributes to a candidate's chances of winning and testing this by comparing candidates' win rates to their spending levels.


3. **Hypothesis 3** (H3): Campaign spending efficiency (measured in terms of expenditure per vote) varies greatly by district and candidate. This hypothesis investigates whether certain districts require more or less spending to gain votes, accounting for changes in voter behavior, media expenses, and other district-specific considerations. By assessing spending efficiency, we hope to find patterns and gain insights into the cost-effectiveness of campaign methods.

These hypotheses will be tested using correlation analysis, regression models, and efficiency estimates. The findings will help to clarify how financial resources influence electoral outcomes and provide practical advice for optimizing campaign spending.

### Research Design

I will use data from two main sources: the Federal Election Commission (FEC) for comprehensive campaign spending and the North Carolina State Board of Elections for official election results. The FEC data contains detailed details of candidate spending, split down into advertising, events, and staff wages. The election results data shows the total number of votes earned by each candidate and variables such as total registered voters and voting turnout in each district. These publicly available datasets provide the information required for detailed investigation.

Before analyzing the data, I'll need to preprocess it to ensure it's appropriate. This includes recognizing and addressing missing values, standardizing data types (e.g., ensuring all financial statistics are in the same currency), and combining databases based on shared identifiers such as candidate ID and district number. This technique will produce a uniform dataset that combines expenditure information with election results.


My research will comprise descriptive statistics to highlight key variables, correlation analysis to evaluate the relationship between spending and vote share, and regression modeling to quantify the influence of the expenditure on vote percentages. I'll also assess spending efficiency by estimating the cost per vote across districts and candidates. This organized approach thoroughly examines the relationship between campaign spending and electoral outcomes, offering useful information to political strategists, candidates, and policymakers.


### Conclusions
My findings show a strong positive correlation between campaign spending and vote percentages in North Carolina congressional races. This means that candidates who spend more money are more likely to win a higher vote share, validating that financial resources are essential for increasing publicity and voter engagement. Regression analysis demonstrates that more spending leads to higher vote percentages, emphasizing the significance of campaign money in election success. The evidence includes statistical analyses that show a significant relationship between spending and voting outcomes. However, acting on this research necessitates that the numbers accurately reflect all relevant spending and that the relationship is causal. It also assumes political stability and that the conclusions apply to similar situations.

The study provides useful insights despite limitations such as potential unobserved factors, a focus on a single state, and a need for more diversity across spending categories. These findings suggest that political campaigns can be more effective by carefully optimizing spending, and legislators should consider the significant influence of money in politics when drafting campaign finance legislation. This report recommends improving campaign strategy and influencing policy debates to support fair and competitive elections.

Additional research may expand on this work by looking into the impact of campaign expenditures in different states and countries, providing a comparative view of the role of money in diverse political systems. Cross-comparison studies with other types of elections, such as gubernatorial, presidential, and local elections, can assist in determining whether the trends shown in North Carolina are consistent across political contexts. These new research directions will aid in our understanding of the intricate interplay between money and politics, resulting in more effective campaign strategies and policy decisions.
